```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Get number of unique schools and start data frame.  Go back to pymoli and decrease coding lines by using this format for starting the database and naming the first column.
dis_sum_df = pd.DataFrame([{"Total Schools":school_data_complete["School ID"].nunique()}])

#Getting the total number of students and then adding it to the dataframe
dis_sum_df["Total Students"]= school_data_complete["Student ID"].count()

#Creating a list of unique schools.  This list drops most students, so do not use it for anything else!!!!!
unique_school_list = school_data_complete.drop_duplicates("School ID", keep="first", inplace=False)
#Use the unique school list to get the total budget
dis_sum_df["Total Budget"] = unique_school_list["budget"].sum()

#Get the average math score per student.  NOTE go back to the original database and do not use the unique school list!!!
dis_sum_df["Average Math Score"] = school_data_complete["math_score"].mean()

#Get the average reading score per student.
dis_sum_df["Average Reading Score"] = school_data_complete['reading_score'].mean()

#create a new list with only students that have passing math scores.  NOTE I CAN USE THESE LATER
math_passing = school_data_complete.loc[school_data_complete["math_score"]>=70, :]
#get the percentage of students with passing math scores.
dis_sum_df["% Passing Math"] = 100 * math_passing["Student ID"].count() / dis_sum_df["Total Students"]

#Create new list with passing reading scores
read_passing = school_data_complete.loc[school_data_complete["reading_score"]>=70, :]
#get the percentage of students with passing reading scores.
dis_sum_df["% Passing Reading"] = 100 * read_passing["Student ID"].count() / dis_sum_df["Total Students"]

#Get the overall passing rate using the formula from the description
dis_sum_df["Overall Passing Rate"] = (dis_sum_df["Average Math Score"] + dis_sum_df["Average Reading Score"])/2

#Map columns
dis_sum_df["Total Students"]=dis_sum_df["Total Students"].map("{:,}".format)
dis_sum_df["Total Budget"]=dis_sum_df["Total Budget"].map("${:,.2f}".format)
dis_sum_df["Average Math Score"]=dis_sum_df["Average Math Score"].map("{:,.2f}".format)
dis_sum_df["Average Reading Score"]=dis_sum_df["Average Reading Score"].map("{:,.2f}".format)
dis_sum_df["% Passing Math"]=dis_sum_df["% Passing Math"].map("{:,.2f}".format)
dis_sum_df["% Passing Reading"]=dis_sum_df["% Passing Reading"].map("{:,.2f}".format)
dis_sum_df["Overall Passing Rate"]=dis_sum_df["Overall Passing Rate"].map("{:,.2f}".format)

dis_sum_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#create a group object of data grouped by school names.  I have alredy checked the School ID and school name columns and there was no discrepancy.
school_group = school_data_complete.groupby(["school_name"])

#Get a list of the school types for each school.
school_type = school_group["type"].first()

#Create a data frame with the first column containing the school types.
school_group_df = pd.DataFrame({"School Type":school_type})

#Fill in columns of dataframe.
school_group_df["Total Students"] = school_group["Student ID"].count()
school_group_df["Total School Budget"] = school_group["budget"].mean()
school_group_df["Per Student Budget"] = school_group_df["Total School Budget"] / school_group_df["Total Students"] 
school_group_df["Average Math Score"] = school_group["math_score"].mean()
school_group_df["Average Reading Score"] = school_group["reading_score"].mean()

#Create groups for passing students for passing percentages.  Refer to passing lists from above.
school_group_math_pass = math_passing.groupby(["school_name"])
school_group_df["% Passing Math"] = 100 * school_group_math_pass["Student ID"].count() / school_group["Student ID"].count()

school_group_reading_pass = read_passing.groupby(["school_name"])
school_group_df["% Passing Reading"] = 100 * school_group_reading_pass["Student ID"].count() / school_group["Student ID"].count()

school_group_df["% Overall Passing Rate"] = (school_group_df["% Passing Math"] + school_group_df["% Passing Reading"]) / 2

#NOTE The school_group_df dataframe will be used later on so you can't map it.  You need to make a NEW dataframe and map that one.
school_group_mapped = pd.DataFrame()

#mapping
school_group_mapped["School Type"] = school_group_df["School Type"]
school_group_mapped["Total Students"] = school_group_df["Total Students"].map("{:,}".format)
school_group_mapped["Total School Budget"] = school_group_df["Total School Budget"].map("${:,.2f}".format)
school_group_mapped["Per Student Budget"] = school_group_df["Per Student Budget"].map("${:,.2f}".format)
school_group_mapped["Average Math Score"] = school_group_df["Average Math Score"].map("{:.2f}".format)
school_group_mapped["Average Reading Score"] = school_group_df["Average Reading Score"].map("{:.2f}".format)
school_group_mapped["% Passing Reading"] = school_group_df["% Passing Reading"].map("{:.2f}".format)
school_group_mapped["% Passing Math"] = school_group_df["% Passing Math"].map("{:.2f}".format)
school_group_mapped["% Overall Passing Rate"] =school_group_df["% Overall Passing Rate"].map("{:.2f}".format)

del school_group_mapped.index.name
school_group_mapped
```

```python
### can use these lists later
#Create lists of students by grade.
grade_9 = school_data_complete.loc[school_data_complete["grade"]=="9th"]
grade_10 = school_data_complete.loc[school_data_complete["grade"]=="10th"]
grade_11 = school_data_complete.loc[school_data_complete["grade"]=="11th"]
grade_12 = school_data_complete.loc[school_data_complete["grade"]=="12th"]


#get averages of math scores, for columns gouped by school name for the dataframe
Mgrade_9_column = grade_9.groupby("school_name")["math_score"].mean()
Mgrade_10_column = grade_10.groupby("school_name")["math_score"].mean()
Mgrade_11_column = grade_11.groupby("school_name")["math_score"].mean()
Mgrade_12_column = grade_12.groupby("school_name")["math_score"].mean()

#create database
Mgrade_df = pd.DataFrame({"9th":Mgrade_9_column, "10th":Mgrade_10_column, "11th":Mgrade_11_column, "12th":Mgrade_12_column})

#clean table
Mgrade_df["9th"] = Mgrade_df["9th"].map("{:.2f}".format)
Mgrade_df["10th"] = Mgrade_df["10th"].map("{:.2f}".format)
Mgrade_df["11th"] = Mgrade_df["11th"].map("{:.2f}".format)
Mgrade_df["12th"] = Mgrade_df["12th"].map("{:.2f}".format)

del Mgrade_df.index.name
Mgrade_df
```
